Detect and Remove outlier

```python
from scipy import stats
from warnings import filterwarnings
filterwarnings('ignore')
```

The Z-score method  help us to find outlier by measuring how far each value is from the average(mean) in term of standard daviation,Z-score tells us how many standard daviation values is away from the mean 


```python
from scipy.stats import zscore
import numpy as np
```

```python
#calculate zscore value for paymenttier and age
z_score=np.abs(zscore(df[['Age','PaymentTier']]))
```

```python
z_score
```

```python
# Identify rows with zscore>3 (Outliers)
outliers=(z_score>3).any(axis=1)
```

```python
outliers
```

```python
#Remove outliers from dataset
df_no_outliers=df[~outliers]
```

```python
df_no_outliers.head()
```
