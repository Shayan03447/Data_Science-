Handling outlier

The Z-score method help us find outlier by measuring how far each value is from the average(mean) in term of standard daviation AZ-score tells us how many standard daviation a value is away from the mean,

If the z score value is greater than 3, the value is considered an outlier because its far from the mean.

```python
from scipy.stats import zscore
import numpy as np

#calculate z-score value for PaymentTier,Age
z_score=np.abs(zscore(df_cleaned[['Age','PaymentTier']]))
```

```python
z_score
```

Identify rows with Z-score > 3 (outlier)

```python
outliers=(z_score>3).any(axis=1)
```

```python
outliers
```

Remove outliers from the dataset

```python
df_no_outliers=df_cleaned[~outliers]
```
