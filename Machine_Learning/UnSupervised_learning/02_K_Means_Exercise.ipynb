ELBOW METHOD TO FIND THE BEST CLUSTER

```python
k_rang=range(1,10)
sse=[]
for k in range(1,10):
    km=KMeans(n_clusters=k)
    km.fit(df[['petal length (cm)','petal width (cm)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.scatter(k_rang,sse)
```
