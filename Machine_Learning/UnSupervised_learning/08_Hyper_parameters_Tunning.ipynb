```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(iris['data'],iris['target'],test_size=0.3)
```

Just assume that you are going to svm model

```python
model=SVC(kernel='rbf',C=30,gamma='auto')
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

The result change in every itration so thats why we use k-fold cross validation 

we have a method of cross-val-score which can tell you the score of each itration here 

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(SVC(kernel='rbf',C=10,gamma='auto'),iris['data'],iris['target'],cv=5)
```

```python
cross_val_score(SVC(kernel='linear',C=10,gamma='auto'),iris['data'],iris['target'],cv=5)
```

```python
cross_val_score(SVC(kernel='rbf',C=20,gamma='auto'),iris['data'],iris['target'],cv=5)#cv means five fold
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf=GridSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
},cv=5,return_train_score=False)#first parameter is your model,second the value of c,and third the value of kernel and at the end cv means cross validation =5
#once it is done you will do the model train 
clf.fit(iris['data'],iris['target'])
clf.cv_results_
```

cv result are not easy to view but sklearn provide a way to download these result into the dataframe 

```python
df=pd.DataFrame(clf.cv_results_)
```

```python
df
```

```python
df[['param_C','param_kernel','mean_test_score']]
```

high computational cost in gridcv to tackle this computational cost we use randomized search cv it will try a random combination of these parameters

```python
from sklearn.model_selection import RandomizedSearchCV
rs=RandomizedSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
},cv=5,
return_train_score=False,
n_iter=2)
```

```python
rs.fit(iris['data'],iris['target'])
rs.cv_results_
```

```python
df=pd.DataFrame(rs.cv_results_)
```

```python
df
```

```python
df[['param_C','param_kernel','mean_test_score']]
```

Hyperparameter Tunning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
model_params={
    'svm':{
        'model':SVC(gamma='auto'),
        'params':{
            'C':[1,10,20],
            'kernel':['rbf','linear']
        }
    },
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10]
        }
    },
    'LogisticRegression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C':[1,5,10]
        }
    }
    
}
```

```python
score=[]

for model_name, mp in model_params.items():
    clf=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(iris['data'],iris['target'])
    score.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
```

```python
df=pd.DataFrame(score,columns=['model','best_score','best_params'])
```
