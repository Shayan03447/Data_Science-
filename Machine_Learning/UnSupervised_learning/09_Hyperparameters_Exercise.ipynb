It can also done with the gridsearchcv

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf=GridSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
},cv=5,return_train_score=False)
#Now train the model
clf.fit(digits['data'],digits['target'])
clf.best_score_
```

in gridsearchsv you must give the parameters 

```python
clf1=GridSearchCV(LogisticRegression(max_iter=1000),{
    "C":[0.1,1,10],
    'solver':['liblinear','saga']
},cv=5,return_train_score=False)
clf1.fit(digits['data'],digits['target'])
clf1.best_score_
```

In gridSearchsv it has high computational cost so we overcome this computational cost by random search cv it is similer to the gridsearchcv

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rf=RandomizedSearchCV(LogisticRegression(max_iter=1000),{
    'C':[1,10,20],
    'solver':['liblinear','saga']
},cv=5,return_train_score=False)
rf.fit(digits['data'],digits['target'])
rf.best_score_
```

if we have more models so we use this type of coding

```python
model_params={
    'LogisticRegression':{
        'model':LogisticRegression(solver="liblinear",multi_class='auto'),
        'params':{
            'C':[1,5,10]
        }
    },
    'SVM':{
        'model':SVC(gamma='auto'),
        'params':{
            'C':[1,10,20],
            'kernel':["rbf",'linear']
        }
    },
    'RandomForestClassifier':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10 ]
        }
    },
    'MultinomialNB':{
        'model':MultinomialNB(alpha=1.0),
        'params':{
            'fit_prior':[True,False]

        }
    },
    'GaussianNB':{
        'model':GaussianNB(),
        'params':{
           # 'priors':[[0.5,0.3,0.2]]
        }
    },
    'Decision_Tree':{
        'model':DecisionTreeClassifier(criterion='entropy'),
        'params':{
            'max_depth':[5,10],
            'min_samples_split':[2,5,10]
        }
    }
}
```

```python
score=[]

for model_name, mp in model_params.items():
    clf=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(digits['data'],digits['target'])
    score.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
```

```python
df=pd.DataFrame(score,columns=['model','best_score','best_params'])
```
