Setting up the data

```python
loans.head()
```

```python
loans['purpose'].value_counts()
```

```python
loans.info()
```

```python
cat_feats=['purpose']
```

purpose is the catagorical feature,so we need to fixed that column in dummy feature for making it the numeric column

```python
final_data=pd.get_dummies(loans,columns=cat_feats,drop_first=True)
```

```python
final_data.info()
```

```python
final_data.head()
```

CROSS VALIDATION(TRAIN_TEST_SPLIT)

```python
from sklearn.model_selection import train_test_split
```

```python
x=final_data.drop('not.fully.paid',axis=1)
y=final_data['not.fully.paid']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
```

Fit the model

```python
dtree.fit(x_train,y_train)
```

Prediction and Evaluation of Decision Tree

```python
prediction=dtree.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,prediction))
print('/n')
print(confusion_matrix(y_test,prediction))
```

TRAINING THE RANDOM FOREST MODEL

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=300)
```

FIT THE MODEL

```python
rfc.fit(x_train,y_train)
```

PREDICTION AND EVALUATION

```python
Random_prediction=rfc.predict(x_test)
```

```python
print(classification_report(y_test,Random_prediction))
print('/n')
print(confusion_matrix(y_test,Random_prediction))
```

Save the model using joblib and pickle:it allows you to serialize python object into your file

```python
import pickle
```

```python
with open('model_pickle','wb') as f:
    pickle.dump(rfc,f)
```

```python
with open('model_pickle','rb') as f:
    mp=pickle.load(f)
```

```python
mp.predict(x_test)
```
