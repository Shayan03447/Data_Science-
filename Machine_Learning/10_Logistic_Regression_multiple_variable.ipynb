from sklearn.datasets import load_digits:that is used for getting ready-made datasets for learning machine learning algorithm

```python
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits['target'][0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=digits['data']
y=digits['target']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
logmodel.score(x_test,y_test)
```

```python
plt.matshow(digits.images[67])
```

```python
digits.target[67]
```

```python
logmodel.predict([digits.data[67]])
```

```python
logmodel.predict(digits.data[0:5])
```

```python
y_prediction=logmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_prediction)
cm
```

lets create the heatmap of this confusion_matrix

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
