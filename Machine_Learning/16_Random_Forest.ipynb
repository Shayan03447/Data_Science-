```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
digits=load_digits()
```

Into DataFrame

```python
df=pd.DataFrame(digits['data'])
```

```python
df.head()
```

```python
digits['target']
```

Append the new column into the dataframe

```python
df['target']=digits['target']
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['target'],axis=1)
y=df['target']
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
len(x_train)
```

```python
len(y_train)
```

```python
len(x_test)
```

```python
len(y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

Now why is it called an ensambles:Is a term used when you are using multiple algorithms to predict the outcomes.

```python
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

Now tune this model by giving 20 Random_Trees

```python
model1=RandomForestClassifier(n_estimators=20)
```

```python
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
y_predict=model.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_predict)
cm
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)
plt.xlabel("prediction")
plt.ylabel("Truth")
```
