THE COLUMN WE HAVE IS CATAGORICAL FOR THE MODEL WE NEED THE DATA IN NUMERIC FOAM CONVERT THE OBJECT COLUMN INTO THE NUMERIC BY DUMMY VARIABLES 

```python
dummies=pd.get_dummies(df['Car Model'],dtype=int)
dummies
```

Merged that two dataframe by concatination

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
merged.info()
```

Whenever one variable is drived from the rest of variables these variables are called multico_linear variabled or when the multi co_linearity in your dataframe it create the problem of dummy variable trap.the thumb of the rule is drop one dummy variable column

```python
final=merged.drop(['Car Model','Audi A5'],axis=1)
final
```

```python
final.info()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
x=final.drop(['Sell Price($)'],axis=1)
x
y=final['Sell Price($)']
```

```python
lr.fit(x,y)
```

```python
lr.predict([[2800,6,0,1]])
```
